```python
from kpis import *
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sales_region = sales_region.reset_index()
sales_category = sales_category.reset_index()
top_10_products = top_10_products.reset_index()
sales_trend = sales_trend.reset_index()
```

```python
sns.barplot(data =sales_region, x="region", y="sales", color="green")

plt.title("sales by region")

plt.show()
```

```python
sns.barplot(data=sales_category, x="category",y="sales", color="blue")

plt.title("sales by category")

plt.show()
```

```python
sns.barplot(data=top_10_products, y="product_name", x="sales", color="yellow")

plt.title("top 10 products")

plt.show()
```

```python
sns.lineplot(data=sales_trend, x="order_date", y="sales")

plt.title("sales trend")

plt.show()
```
